## Metric Evaluation

Evaluasi metrik yang akan digunakan merupakan evaluasi untuk Machine Learning Regresi, yang digunakan untuk mengukur kinerja model terhadap fitur-fitur yang nanti akan ditentukan. Metrik evaluasi yang digunakan ialah MAE, RMSE, MAPE, dan R-squared


- Mean Absolute Error (`MAE`):
  *   MAE merepresentasikan perbedaan rata - rata absolut antara nilai prediksi dengan nilai aktual.
  *   Nilai MAE yang kecil mengindikasikan secara rata - rata, prediksi model mendekati nilai aktual. Pada konteks project ini, matriks MAE sangat penting karena berhubungan langsung dengan goal dari project, yaitu memprediksi harga secara akurat.
- Root Mean Square Error (`RMSE`):
  *   RMSE adalah akar perbedaan rata - rata antara nilai prediksi dengan nilai aktual.
  *   Nilai RMSE yang kecil mengindikasikan secara rata - rata, prediksi model mendekati nilai aktual. Pada konteks project ini, matriks MAE sangat penting karena berhubungan langsung dengan goal dari project, yaitu memprediksi harga secara akurat.
- Mean Absolute Percentage Error (`MAPE`):
  *   MAPE adalah persentase perbedaan rata - rata absolut antara nilai prediksi dengan nilai aktual.
  *   MAPE memberikan informasi seberapa besar kesalahan prediksi model dibandingkan dengan nilai aktual. Semakin kecil nilai presentasi kesalahan (percentage error) pada MAPE maka semakin akurat hasil prediksi suatu model.
- R-squared (`R2`):
  *   R2 mengukur proporsi varians pada variabel target yang dijelaskan oleh model.
  *   R2 yang tinggi menunjukkan sebagian besar variabilitas harga mobil ditangkap oleh model. Pada konteks project ini, matriks R2 penting karena menunjukkan bahwa model secara efektif memanfaatkan fitur yang disediakan untuk memprediksi harga.

File-file CSV diatas akan digabungkan menjadi satu data frame, dengan dilakukan beberapa penyesuaian yaitu :
1. penyamaan nama kolom (`tax()` menjadi `tax`),
1. penambahan kolom `brand` untuk membedakan masing-masing merek dan file CSV,

```python
# List file CSV
csv_files = {
     'audi.csv': 'Audi',
     'bmw.csv': 'BMW',
     'ford.csv': 'Ford',
     'hyundi.csv': 'Hyundai',
     'merc.csv': 'Mercedes',
     'skoda.csv': 'Skoda',
     'toyota.csv': 'Toyota',
     'vauxhall.csv': 'Vauxhall',
     'vw.csv': 'Volkswagen'
     }

# membuat dataframe untuk penggabungan
merged_df = pd.DataFrame()

# menggabungkan dan membuat beberapa penyesuaiian
for file, brand_name in csv_files.items():
    df_file = pd.read_csv(file)

    if 'tax()' in df_file.columns:
        df_file.rename(columns={'tax()': 'tax'}, inplace=True)

    df_file.insert(0, 'brand', brand_name, allow_duplicates=True)
    df_file['brand'] = df_file['brand'].astype(str)

    merged_df = pd.concat([merged_df, df_file], ignore_index=True)

if 'price' in merged_df.columns:
    price_column = merged_df.pop('price')
    merged_df['price'] = price_column
```

## Exploratory Data Analysis (EDA) for Data Cleaning

### Normality Test
menggunakan metode tes normalitas D'Agostino dan Pearson's test untuk menguji normalitas data numeric

```python
# cek normalitas D'Agostino pada data numeric
num_ver = ['mileage', 'tax', 'mpg', 'engineSize', 'year']
for i in num_ver:
    dp_statistic, dp_pvalue = normaltest(merged_df[i])

    if dp_pvalue > 0.05:
        print(f'P-value for {i}: {dp_pvalue}. Normal')
    else:
        print(f'P-value for {i}: {dp_pvalue}. Not normal')
```

pada uji normalitas diatas, terhitung bahwa sebaran data pada fitur-fitur numerikal tidaklah normal, untuk lebih jelasnya akan dapat dilihat pada sebaran data dengan plot dibawah nanti.

### Corelation Matrix

berikutnya kita akan melihat hubungan korelasi antar fitur yang akan digambarkan dalam *Corelation Matrix*

```python
# Define kolom-kolom numerik
numerical_columns = ['year', 'mileage', 'mpg', 'engineSize']
numerical_df = merged_df[numerical_columns]

# Membuat Correlation matrix
plt.figure(figsize=(8, 8))
palette = sns.diverging_palette(20, 220, n=256)
corr = numerical_df.corr(method='spearman')
sns.heatmap(corr, annot=True, fmt='.2f', cmap=palette, square=True, linewidths=.5)
plt.title('Correlation Matrix', size=15, weight='bold')
plt.show()
```

Correlation matrix menunjukkan ada fitur yang memiliki nilai korelasi yang tinggi yaitu fitur `year` dan `mileage`, dimana seharusnya fitur-fitur yang memiliki nilai korelasi tinggi hanya akan dipakai salah satunya saja, dan sisanya akan didrop. Hal ini dikarenakan:
- `Masalah Kolinearitas`: Dalam beberapa kasus, fitur yang sangat berkorelasi dapat menyebabkan masalah multikolinearitas, khususnya pada model linier. Multikolinearitas terjadi ketika dua atau lebih variabel independen dalam suatu model regresi berkorelasi tinggi sehingga sulit untuk menentukan pengaruh individual masing-masing variabel terhadap variabel dependen.
- `Efisiensi Komputasi`: Menyertakan fitur yang sangat berkorelasi dapat memperlambat proses pelatihan dan meningkatkan biaya komputasi model.

### Check whitespace pada dataset

```python
brand_prices = train_data.groupby('brand')['price'].agg(['sum', 'count']).reset_index()
brand_prices.columns = ['brand', 'sum_price', 'count']
brand_prices['average_price'] = (brand_prices['sum_price'] / brand_prices['count']).round(2)
brand_prices['percentage_sales'] = (brand_prices['sum_price'] / brand_prices['sum_price'].sum()) * 100
brand_prices['percentage_sales'] = brand_prices['percentage_sales'].round(2)

brand_prices = brand_prices.sort_values(by='sum_price', ascending=False).reset_index(drop=True)

brand_prices
```

```python
plt.figure(figsize=(10, 6))
brand_colors = {brand: sns.color_palette('deep')[i] for i, brand in enumerate(train_data['brand'].unique())}
barplot = sns.barplot(data=brand_prices, x='brand', y='sum_price', palette=brand_colors)
plt.title('Sum of Prices by Brand')
plt.xlabel('Brand')
plt.ylabel('Sum of Price')

for index, row in brand_prices.iterrows():
    barplot.text(index, row['sum_price'], f"{row['sum_price']/1000000:.0f}M ", color='black', ha='center', va='bottom')
    barplot.text(index, row['sum_price']*0.5, f"{row['percentage_sales']}%", color='black', ha='center', va='center')


plt.show()
```

## Preprocessing

```python
# pipeline ordinal data
merged_df_copy = train_data.copy()
nom_ordinal = ['year', 'brand', 'model', 'fuelType', 'transmission']

sorted_brand = merged_df_copy.groupby('brand')['price'].mean().sort_values().index.tolist()
sorted_model = merged_df_copy.groupby('model')['price'].mean().sort_values().index.tolist()
sorted_fueltype = merged_df_copy.groupby('fuelType')['price'].mean().sort_values().index.tolist()
sorted_transmission = merged_df_copy.groupby('transmission')['price'].mean().sort_values().index.tolist()
year_range = list(range(2010, 2021))

categories = [year_range, sorted_brand, sorted_model, sorted_fueltype, sorted_transmission]

ordinal_pipeline = Pipeline([
    ('ordinal_encode', OrdinalEncoder(categories=categories))
])

# pipeline numeric data
num_scaler = ['tax', 'mpg', 'engineSize']
scaler = RobustScaler()
numerical_pipeline = Pipeline([
    ('scaler', scaler)
])
```

Dalam case ini data categoric akan di encoding menggunakan metode `ordinal encoding`, sedangkan data numeric akan di-scaled menggunakan `robust scaler`. <br/>

**Ordinal encoder:** <br/>
 - Transform feature `year` dengan metode ordinal encoding, karena feature ini mengandung nilai tingkatan/hierarki yang bermakna. Dalam case ini feature tahun memiliki pengaruh terhadap harga used car yang mana semakin baru tahun registrasi semakin mahal pula harga untuk used car dan sebaliknya. Ini memungkinkan model untuk memahami urutan waktu dan memperlakukan feature tahun dengan benar dalam proses modeling nanti. <br/>
 - Transform feature `model` dengan metode ordinal encoding berdasarkan rata-rata harga, guna memberikan informasi hierarki pada tiap model mobil agar model machine learning mampu memprediksi harga dengan lebih tepat.
 - Transform feature `brand` dengan metode ordinal encoding berdasarkan rata-rata harga, guna memberikan informasi hierarki pada tiap brand mobil agar model machine learning mampu memprediksi harga dengan lebih tepat.
 - Transform feature `fuelType` dengan metode ordinal encoding berdasarkan rata-rata harga, guna memberikan informasi hierarki pada tiap tipe bahan bakar mobil agar model machine learning mampu memprediksi harga dengan lebih tepat.
 - Transform feature `transmission` dengan metode ordinal encoding berdasarkan rata-rata harga, guna memberikan informasi hierarki pada tiap jenis transmisi mobil agar model machine learning mampu memprediksi harga dengan lebih tepat.

**Robust scaler:** <br/>
 Scalled feature `tax`, `mpg`, dan `engineSize` dengan robust scaler, karena feature tersebut memiliki distribusi yang tidak normal/skew, sehingga robust scaler dipilih karena  lebih tahan terhadap outlier.

```python
# column transformer semua pipeline
preprocessor = ColumnTransformer([
    ('numeric', numerical_pipeline, num_scaler),
    ('ordinal', ordinal_pipeline, nom_ordinal)
])

preprocessor.fit(X_train)

numeric_feature_names = num_scaler
ordinal_feature_names = nom_ordinal

feature_names = numeric_feature_names + ordinal_feature_names

X_train_trans = pd.DataFrame(preprocessor.transform(X_train), columns=feature_names)
X_test_trans = pd.DataFrame(preprocessor.transform(X_test), columns=feature_names)
```

## Modelling

```python
# baseline models
models = [
    ['Linear Regression', LinearRegression()],
    ['KNN', KNeighborsRegressor()],
    ['DecisionTree', DecisionTreeRegressor(random_state=42)],
    ['RandomForest', RandomForestRegressor(random_state=42)],
    ['XGBoost', XGBRegressor(random_state=42)]
]
```

```python
# fungsi untuk training score
def training_score(models, X_train, y_train):
    model_name = []
    training_r2 = []
    training_rmse = []
    training_mae = []

    for name, model in models:
        pipeline = Pipeline([
            ('preprocessor', preprocessor),
            ('model', model)
        ])
        model_name.append(name)

        # training proses
        pipeline.fit(X_train, y_train)
        y_pred = pipeline.predict(X_train)
        training_r2.append(r2_score(y_train, y_pred))
        training_rmse.append(np.sqrt(mean_squared_error(y_train, y_pred)))
        training_mae.append(mean_absolute_error(y_train, y_pred))

    return pd.DataFrame({
        'Model': model_name,
        'Training R2': training_r2,
        'Training RMSE': training_rmse,
        'Training MAE': training_mae
    })

# fungsi untuk validation score
def validation_score(models, X_train, y_train):
    model_name = []
    cv_r2_mean = []
    cv_r2_std = []
    cv_rmse_mean = []
    cv_rmse_std = []
    cv_mae_mean = []
    cv_mae_std = []
    cv_mape_mean = []
    cv_mape_std = []

    for name, model in models:
        pipeline = Pipeline([
            ('preprocessor', preprocessor),
            ('model', model)
        ])
        model_name.append(name)

        # set scoring
        scoring= ['r2', 'neg_root_mean_squared_error', 'neg_mean_absolute_error']

        # validation
        cv_score = cross_validate(pipeline, X_train, y_train, scoring=scoring, cv=5, n_jobs=-1)

        cv_r2_mean.append(cv_score['test_r2'].mean())
        cv_r2_std.append(cv_score['test_r2'].std())

        cv_rmse_mean.append(abs(cv_score['test_neg_root_mean_squared_error']).mean())
        cv_rmse_std.append(abs(cv_score['test_neg_root_mean_squared_error']).std())

        cv_mae_mean.append(abs(cv_score['test_neg_mean_absolute_error']).mean())
        cv_mae_std.append(abs(cv_score['test_neg_mean_absolute_error']).std())

        y_true = y_train
        y_pred_cv = cross_val_predict(pipeline, X_train, y_train, cv=5)
        mape_values_cv = np.abs((y_true - y_pred_cv) / y_true) * 100
        cv_mape_mean.append(mape_values_cv.mean())
        cv_mape_std.append(mape_values_cv.std())

    return pd.DataFrame({
        'Model': model_name,
        'CV R2 (mean)': cv_r2_mean,
        'CV R2 (std)' : cv_r2_std,
        'CV RMSE (mean)': cv_rmse_mean,
        'CV RMSE (std)': cv_rmse_std,
        'CV MAE (mean)' : cv_mae_mean,
        'CV MAE (std)' : cv_mae_std,
        'CV MAPE (mean)': cv_mape_mean,
        'CV MAPE (std)': cv_mape_std
    })

train_cv = training_score(models, X_train, y_train)
print('Training CV score:')
display(train_cv)

val_cv = validation_score(models, X_train, y_train)
print('\nValidation CV score:')
display(val_cv)
```

Model Random Forest jika dibandingkan dengan model lain memiliki score R2 yang tinggi dan score RMSE, MAE, dan MAPE yang paling kecil. Oleh karena itu, Random Forest akan dipilih sebagai final model.

**Hyperparameter Tuning (Gridsearch)**

Terlihat tidak ada peningkatan score R2 setelah dilakukan parameter tunning, oleh karena itu model Random Forest tanpa parameter tunning akan dipilih sebagai final model.
